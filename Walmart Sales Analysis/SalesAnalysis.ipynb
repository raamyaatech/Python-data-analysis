**EDA**

**what is the monthly sale trend?**

```python
sales["order_date"].min()
```

```python
sales["order_date"].max()
```

**Add month_year columns**

```python
sales["month_year"] = sales["order_date"].apply(lambda x: x.strftime("%Y-%m"))
sales["month_year"]
```

**Monthly trends**

```python
sales_trend = sales.groupby(["month_year"])["sales"].sum().reset_index()
                                            
                                            
```

```python
sales_trend
```

**plot the monthly sales trend**

```python
plt.figure(figsize = (18,6))
plt.plot(sales["month_year"], sales["sales"])
plt.xticks(rotation='vertical',size =  8)
plt.show()
```

**What are the top 10 products by Sales?**

```python
product_sales = sales.groupby(["product_name"])["sales"].sum().reset_index().sort_values(["sales"] , ascending = False)
```

```python
product_sales.iloc[:10]
```

**what are the most selling products?**

```python
most_selling_product = sales.groupby(["product_name"])["quantity"].sum().reset_index().sort_values(["quantity"],ascending = False)
```

```python
most_selling_product
```

**what are the most preferred ship Mode?**

```python
plt.figure(figsize = (8,8))
sns.countplot(data = sales, x = "ship_mode")
```

**which are the most preferrable Category and Sub Category?**

```python
category_profits = sales.groupby(["category", "sub_category"])["profit"].sum().reset_index().sort_values(["category","sub_category"],ascending = False)
```
